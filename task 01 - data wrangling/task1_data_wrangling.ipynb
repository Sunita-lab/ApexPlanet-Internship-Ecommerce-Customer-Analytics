## Dataset Upload
The dataset is uploaded to the project directory under `data/customer.xlsx`. It contains transactional data from an online retail store, including details such as InvoiceNo, StockCode, Description, Quantity, InvoiceDate, UnitPrice, CustomerID, and Country.

```python
import pandas as pd
df = pd.read_excel("../data/customer.xlsx")
df.shape
```

# Data Cleaning

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
df = df.dropna(subset=["Description"])
```

```python
df.shape
```

```python
df.isnull().sum()
```

# Data Transformation

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
```

InvoiceDate column was converted to datetime format to support future time-based analysis.

# Export Cleaned Dataset

```python
df.to_csv("cleaned_online_retail.csv", index=False)
```
